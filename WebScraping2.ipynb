```python
import requests
from lxml import etree
from selenium.webdriver import Chrome
from selenium.webdriver.support.select import Select
from selenium.webdriver.chrome.options import Options #Don't show web page window
import time
from bs4 import BeautifulSoup
from pygtrans import Translate


#不显示浏览器使用的过程
opt = Options()
opt.add_argument('--headless')
opt.add_argument('--disable-gpu')
opt.add_argument('start-maximized')
#-------------------------------------
web = Chrome(options=opt,executable_path='./chromedriver')
web.maximize_window()
url = "https://wallstreetcn.com/live/us-stock"


web.get(url)
time.sleep(5)
text = web.page_source
soup=BeautifulSoup(text,"html.parser")
news_list=soup.select('p')

client = Translate()


for i in news_list:
    print(i.get_text())
    print(' ')


web.quit()
```
