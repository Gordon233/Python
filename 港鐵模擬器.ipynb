# 模擬一個人坐地鐵往返[a,b,c,d]四個地方

```python
from random import randint


class person():
    def __init__(self,name,location):
        self.__name = name
        self.__location = location #一開始所在地
        self.__transportation = None
        
    def __str__(self):
        return f"姓名:{self.__name},位於：{self.__location},八達通餘額：{self.__octopus}."
    
    @property
    def name(self):
        return self.__name
    
    @property
    def location(self):
        return self.__location
    
#     @location.setter
#     def change_location(self,new_location)：
#         self.__location = new_location
    @property
    def transportation(self):
        return self.transportation
    
    @transportation.setter
    def transportation_service(self,name):
        self.__transportation = name
    
    def goto(self,target):
        is_arrive = self.__transportation.deliver(target,self.__location)
        if is_arrive == True:
            self.__location = target
        
    
    
class MTR():
    def __init__(self):
        self.__location=["a","b","c","d"]
        self.__octopus = 0
        
    @property
    def octopus(self):
        return self.__octopus
    
#     @octopus.setter
#     def add_value(self,num):
#         self.__octopus += num 
    
    def deliver(self,where,origin):
        fee = randint(10,30) #隨機生成車資
        if where == origin:
            print("已在此地，無需乘搭")
            return False 
        elif where not in self.__location:
            print("沒有此地點，請重新輸入")
            return False  
        elif where in self.__location and fee<self.__octopus:
            print(f"車資:{fee}")
            self.__octopus -=fee
            print(f"已到達{where},餘額為{self.__octopus}")
            return True
        elif where in self.__location and fee>self.__octopus:
            print(f"車資:{fee}")
            while True:
                try:
                    num = float(input("餘額不足請充值，請先輸入充值金額："))
                    if isinstance(num,float) :
                         break
                except:
                        print("請輸入數字")
            self.__octopus +=num   
            print(f"已充值，目前餘額為{self.__octopus}")
            return False
          
        

class main():
    Tom = person("Tom","a")
    subway = MTR()
    Tom.transportation_service=subway
    
    while True:
        destination = input("從以下四個地方選擇你的目的地，[a,b,c,d]")
        print(f"你選擇去{destination}")
        Tom.goto(destination)
   

if __name__ == "__main__":
    main()
```
